## Data cleaning

```python
feature_na = [i for i in data.columns if data[i].isnull().sum() > 0]
feature_na
```

```python
for i in feature_na:
    print(f"{i} has {np.round((data[i].isnull().sum()/len(data[i])*100),4)}% null values")
```

```python
data.Votes.unique()
```

```python
data.dropna(subset=['Votes'],axis=0,inplace=True)
```

```python
def split(x):
    return x.split('/')[0].strip()
```

```python
data['Votes'] = data['Votes'].apply(split)
```

```python
data['Votes']
```

```python
data['rate']
```

```python
data.Votes.unique()
```

```python
data['Votes'].replace(['NEW','-'],0,inplace=True)
```

```python
data.info()
```

```python
data['rate'] = data['rate'].astype(float)
```

```python
data.info()
```

## calculate avg rating for each restaurant

```python
rating = pd.pivot_table(data,index='Restaurant Name',values='Aggregate rating')
rating
```

```python
rating=rating.sort_values(['Aggregate rating'],ascending=False)
rating[0:15]
```

```python
plt.figure(figsize=(15, 8))

sns.barplot(x=rating[0:20]['Aggregate rating'], y=rating[0:20].index, orient="h")

plt.xlabel('Aggregate Rating')
plt.ylabel('Index')
plt.title('Top 20 Aggregate Ratings')
plt.show()
```

We can see that the top resturants have the highest number of average ratings

## distribution of ratings

```python
sns.set_style('whitegrid')
sns.distplot(data['Aggregate rating'])

plt.show()
```

This is not a normal distibution as we can see a spike in '0's i.e the new resturants

```python
from scipy.stats import normaltest

DataToTest = data['Aggregate rating']

stat,p = normaltest(DataToTest)

print("stat=%0.2f,p=%0.30f" % (stat,p))

if p>0.05:
    print("Normal distribution")
else:
    print("Not a normal distribution")
```

### Let's analyse the highly rated restaurants and worst restaurants based on number of ratings given

```python
data.head(3)
```

```python
sns.distplot(data['Votes'])

plt.show()
```

```python
voting_rating = data.groupby('Restaurant Name')[['Votes']].mean().sort_values('Votes',ascending=False)
voting_rating
```

```python
voting_rating['Restaurant Name'] = voting_rating.index
```

```python
voting_rating = voting_rating.reset_index(drop=True)

voting_rating.head()
```

```python
voting_rating = pd.merge(voting_rating,data[['Aggregate rating','Restaurant Name']])
voting_rating
```

```python
vote_top = voting_rating[voting_rating['Votes'] > 5000]
vote_top.head()
```

```python
vote_top = vote_top.groupby('Restaurant Name')[['Aggregate rating']].mean().sort_values('Aggregate rating',ascending=False)
vote_top
```

### highest Rated based on most number of votes

```python

# Ensure that vote_top DataFrame exists and has 'Aggregate rating' and index properly set

plt.figure(figsize=(20, 8), dpi=100)

# Plot the bar chart
sns.barplot(x=vote_top['Aggregate rating'], y=vote_top.index, orient='h')

plt.xlabel('Aggregate Rating')
plt.ylabel('Index')
plt.title('Top Aggregated Ratings')
plt.show()
```

'Byg Brewski Brewing Company' is the most voted resturant with the average rating of 4.9

```python
bad_count = voting_rating[(voting_rating['Aggregate rating'] < 3) & (voting_rating['Aggregate rating'] > 0)]
bad_count
```

```python
bad_resturants = bad_count[bad_count['Votes'] > 500].groupby('Restaurant Name')[['Aggregate rating']].mean()
bad_resturants
```

### worst restaurants based on highest number of votes given

```python
plt.figure(figsize=(20,8))

sns.barplot(x= bad_resturants['Aggregate rating'],y=bad_resturants.index)

plt.show()
```

These resturants have good votes given by the consumer but less rating..which means there is something wrong with these resturants
